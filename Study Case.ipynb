## For read the file and see the frist rows of file.

```r
# read in that data
##install.packages(readr)
library(readr)
dailyActivityData <- read_csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv")
sleepDayData <- read_csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/sleepDay_merged.csv")
# print the first ten rows of our dataframe
head(dailyActivityData)
head(sleepDayData)
```

## Process of clean and organize the data

```r
skim_without_charts(dailyActivityData)
skim_without_charts(sleepDayData)
```

```r
glimpse(dailyActivityData)
glimpse(sleepDayData)
```

```r
activityData <- dailyActivityData %>% select(Id, ActivityDate,TotalDistance, Calories)
head(activityData)
```

```r
sleepDayData<-separate(sleepDayData, SleepDay, into=c('sleepDate', 'time', 'timeDay'), sep=' ')
sleepDayData<-sleepDayData %>% select('Id', 'sleepDate', 'TotalMinutesAsleep')
sleepDayData<-rename(sleepDayData,'ActivityDate'='sleepDate')
head(sleepDayData)
```

```r
##merge two data frames by SleepDay and ActivityDate
newData<-merge(activityData,sleepDayData,by=c('Id', "ActivityDate"))
head(newData)
```

## Data analysis and visualization 

```r
ggplot(newData) + geom_bar(aes(x=TotalDistance, y=TotalMinutesAsleep), stat = "identity") + labs(x = "Total Distance",y = "Sleep Time",
    title = "Relationship between sleep and physical activity", caption = "Dataset: fitbit")
```
